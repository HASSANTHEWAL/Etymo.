```python
# ===============================
# 1. SETUP
# ===============================

!pip install folium

import pandas as pd
import numpy as np
import folium


# ===============================
# 2. LOAD REAL DATA
# ===============================

url = "https://raw.githubusercontent.com/datasets/population/master/data/population.csv"
data = pd.read_csv(url)

# Focus on Yemen
yemen = data[data["Country Name"] == "Yemen, Rep."].copy()


# ===============================
# 3. CREATE RISK LOGIC
# ===============================

# Simulated conflict intensity (placeholder for real conflict data)
yemen["conflict_intensity"] = np.random.randint(0, 10, len(yemen))

# Mine Risk Index (your core idea)
yemen["mine_risk_index"] = (
    (yemen["conflict_intensity"] > 6) &
    (yemen["Value"] > 6000000)
).astype(int)


# ===============================
# 4. CREATE REALISTIC MAP
# ===============================

m = folium.Map(location=[15, 46], zoom_start=6)

for _, row in yemen.iterrows():

    # create clusters (not square)
    for _ in range(25):

        lat = np.random.normal(15, 1.2)
        lon = np.random.normal(47, 2)

        # filter unrealistic points (ocean/outside)
        if lat < 12 or lat > 18:
            continue
        if lon < 42 or lon > 54:
            continue
        if lat < 14 and lon < 47:
            continue

        folium.CircleMarker(
            location=[lat, lon],
            radius=3,
            fill=True,
            fill_color='red' if row["mine_risk_index"] == 1 else 'yellow',
            fill_opacity=0.3
        ).add_to(m)

m
```
